```python
import pandas as pd

dataframe_sampah = pd.read_excel('data_jabar.xlsx', sheet_name='data')
dataframe_sampah
```

```python
#No. 2
tahun = 2020
jumlah = 0

for index, row in dataframe_sampah.iterrows():
    if row['tahun'] == tahun:
        jumlah = jumlah + row['jumlah_produksi_sampah']

print(f"Total Produksi Sampah Pada Tahun 2020 Adalah {jumlah:.2f} ton")
```

```python
#No.3
total_pertahun = {}

for i, row in dataframe_sampah.iterrows():
    tahun = row['tahun']
    jumlah = row['jumlah_produksi_sampah']
    
    if pd.isna(jumlah):
        jumlah = 0

    total_pertahun[tahun] = total_pertahun.get(tahun, 0) + jumlah

print('Total Produksi Sampah Pertahun :')
for thn, jmlh in total_pertahun.items():
    print(f'Tahun {thn} : {jmlh:.2f} ton')
```

```python
#No.4
total_kotakabupaten_pertahun = {}

for i, row in dataframe_sampah.iterrows():
    daerah = row['nama_kabupaten_kota']
    jumlah = row['jumlah_produksi_sampah']
    
    if pd.isna(jumlah):
        jumlah = 0

    total_kotakabupaten_pertahun[daerah] = total_kotakabupaten_pertahun.get(daerah, 0) + jumlah

print('Total Produksi Sampah PerKota/Kabupaten Pertahun :')
for dae, jmlh in total_kotakabupaten_pertahun.items():
    print(f'Tahun {dae} : {jmlh:.2f} ton')
```

```python
dataframe_sampah.to_csv('data_produksi_sampah_jabar.csv', index=False)
dataframe_sampah.to_excel('data_produksi_sampah_jabar.xlsx', index=False)
```
